```python
import pyfolio as pf
import pandas as pd
import datetime as dt

daily_returns_rp = pd.Series(pd.read_csv(r'Output/test_rp_daily_returns.csv')['Returns'])
daily_returns_rp.index = pd.to_datetime(pd.read_csv(r'Output\test_rp_daily_returns.csv')['Date'])
daily_returns_rp = pd.Series(daily_returns_rp)
#print(daily_returns_rp)
```

```python
daily_sp500 = pd.Series(pd.read_csv(r'Data\SP500_index_daily_returns.csv')['SP_500'])
daily_sp500.index = pd.to_datetime(pd.read_csv(r'Data\SP500_index_daily_returns.csv')['Date'])
daily_sp500 = daily_sp500[daily_returns_rp.index[0]:]
#print(daily_sp500)
```

```python
# RP returns
pf.create_returns_tear_sheet(daily_returns_rp, benchmark_rets=daily_sp500)
```

```python
#FRP returns
daily_returns_frp = pd.Series(pd.read_csv(r'Output/test_frp_daily_returns.csv')['Returns'])
daily_returns_frp.index = pd.to_datetime(pd.read_csv(r'Output\test_frp_daily_returns.csv')['Date'])
#print(daily_returns_frp)
pf.create_returns_tear_sheet(daily_returns_frp, benchmark_rets=daily_sp500)
```
