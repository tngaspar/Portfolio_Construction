```python
import pyfolio as pf
import pandas as pd
import datetime as dt

daily_returns_rp = pd.Series(pd.read_csv(r'Output/rp_daily_returns.csv')['Returns'])
daily_returns_rp.index = pd.to_datetime(pd.read_csv(r'Output\rp_daily_returns.csv')['Date'])
daily_returns_rp = pd.Series(daily_returns_rp)
#print(daily_returns_rp)

#FRP returns
daily_returns_frp = pd.Series(pd.read_csv(r'Output/frp_daily_returns.csv')['Returns'])
daily_returns_frp.index = pd.to_datetime(pd.read_csv(r'Output\frp_daily_returns.csv')['Date'])
```

```python
daily_sp500 = pd.Series(pd.read_csv(r'Data\SP500_index_daily_returns.csv')['SP_500'])
daily_sp500.index = pd.to_datetime(pd.read_csv(r'Data\SP500_index_daily_returns.csv')['Date'])
daily_sp500 = daily_sp500[daily_returns_frp.index[0]:]
#print(daily_sp500)
```

```python
# RP returns
pf.create_returns_tear_sheet(daily_returns_rp, benchmark_rets=daily_sp500)
```

```python
#print(daily_returns_frp)
pf.create_returns_tear_sheet(daily_returns_frp, benchmark_rets=daily_sp500)
```

```python
#SP500 returns
sp500 = pd.Series(pd.read_csv(r'Data\SP500_index_daily_returns.csv')['SP_500'])
sp500.index = pd.to_datetime(pd.read_csv(r'Data\SP500_index_daily_returns.csv')['Date'])
sp500 = sp500[dt.datetime(1990,1,1):]
#print(daily_sp500)
pf.create_full_tear_sheet(sp500)
```
